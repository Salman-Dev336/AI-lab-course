## Create pie chart of monthly sum of 'close'

### Subtask:
Generate a pie chart using the `monthly_close_sum` data to visualize the distribution of the sum of the 'close' column across different months.

**Reasoning**:
Create a pie chart to visualize the proportion of the total 'close' sum contributed by each month.

```python
plt.figure(figsize=(10, 10))
plt.pie(monthly_close_sum, labels=monthly_close_sum.index, autopct='%1.1f%%', startangle=140)
plt.title('Monthly Distribution of Total Close Price')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

## Group by month and sum 'close'

### Subtask:
Group the DataFrame by the 'month' column and calculate the sum of the 'close' column for each month.

**Reasoning**:
Group the data by month to aggregate the 'close' prices for each month.

```python
monthly_close_sum = df.groupby('month')['close'].sum()
display(monthly_close_sum)
```

## Convert 'date' to datetime

### Subtask:
Convert the 'date' column to datetime objects to enable easy extraction of month information for subsequent analysis.


**Reasoning**:
Convert the 'date' column to datetime objects for easier manipulation and extraction of month information.



```python
df['date'] = pd.to_datetime(df['date'])
df.info()
```

## Extract month

### Subtask:
Extract the month from the 'date' column and create a new column named 'month' to store this information.


**Reasoning**:
Extract the month from the 'date' column and create a new column named 'month'.



```python
df['month'] = df['date'].dt.month
df.head()
```
