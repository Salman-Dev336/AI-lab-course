BFS

```python
from collections import deque

# Function to print the puzzle grid
def print_puzzle(state):
    for i in range(0, 9, 3):
        print(state[i:i+3])
    print()

# Function to get all possible next moves
def get_moves(state):
    moves = []
    index = state.index(0)
    row, col = divmod(index, 3)

    directions = {
        'up': (row - 1, col),
        'down': (row + 1, col),
        'left': (row, col - 1),
        'right': (row, col + 1)
    }

    for r, c in directions.values():
        if 0 <= r < 3 and 0 <= c < 3:
            new_index = r * 3 + c
            new_state = state[:]
            new_state[index], new_state[new_index] = new_state[new_index], new_state[index]
            moves.append(new_state)
    return moves

# BFS algorithm
def bfs(start, goal):
    queue = deque([[start]])
    visited = set()

    while queue:
        path = queue.popleft()
        state = path[-1]

        if state == goal:
            return path

        visited.add(tuple(state))

        for move in get_moves(state):
            if tuple(move) not in visited:
                queue.append(path + [move])

# ---- MAIN PROGRAM ----
start_state = [1, 2, 3,
               4, 0, 6,
               7, 5, 8]

goal_state = [1, 2, 3,
              4, 5, 6,
              7, 8, 0]

print("=== BFS (Breadth-First Search) ===")
print("Start State:")
print_puzzle(start_state)
print("Goal State:")
print_puzzle(goal_state)

result = bfs(start_state, goal_state)

print("Number of moves:", len(result)-1)
print("Sequence of moves:")
for step in result:
    print_puzzle(step)
```

```python
# Function to print the puzzle grid
def print_puzzle(state):
    for i in range(0, 9, 3):
        print(state[i:i+3])
    print()

# Function to get possible next states
def get_moves(state):
    moves = []
    index = state.index(0)
    row, col = divmod(index, 3)

    directions = {
        'up': (row - 1, col),
        'down': (row + 1, col),
        'left': (row, col - 1),
        'right': (row, col + 1)
    }

    for r, c in directions.values():
        if 0 <= r < 3 and 0 <= c < 3:
            new_index = r * 3 + c
            new_state = state[:]
            new_state[index], new_state[new_index] = new_state[new_index], new_state[index]
            moves.append(new_state)
    return moves

# DFS algorithm
def dfs(start, goal):
    stack = [[start]]
    visited = set()

    while stack:
        path = stack.pop()
        state = path[-1]

        if state == goal:
            return path

        visited.add(tuple(state))

        for move in get_moves(state):
            if tuple(move) not in visited:
                stack.append(path + [move])

# ---- MAIN PROGRAM ----
start_state = [1, 2, 3,
               4, 0, 6,
               7, 5, 8]

goal_state = [1, 2, 3,
              4, 5, 6,
              7, 8, 0]

print("=== DFS (Depth-First Search) ===")
print("Start State:")
print_puzzle(start_state)
print("Goal State:")
print_puzzle(goal_state)

result = dfs(start_state, goal_state)

print("Number of moves:", len(result)-1)
print("Sequence of moves:")
for step in result:
    print_puzzle(step)
```

```python
import heapq

# Function to print the puzzle neatly
def print_puzzle(state):
    for i in range(0, 9, 3):
        print(state[i:i+3])
    print()

# Function to get possible next moves
def get_moves(state):
    moves = []
    index = state.index(0)
    row, col = divmod(index, 3)

    directions = {
        'up': (row - 1, col),
        'down': (row + 1, col),
        'left': (row, col - 1),
        'right': (row, col + 1)
    }

    for r, c in directions.values():
        if 0 <= r < 3 and 0 <= c < 3:
            new_index = r * 3 + c
            new_state = state[:]
            new_state[index], new_state[new_index] = new_state[new_index], new_state[index]
            moves.append(new_state)
    return moves

# Manhattan Distance heuristic
def manhattan_distance(state, goal):
    distance = 0
    for i in range(1, 9):
        x1, y1 = divmod(state.index(i), 3)
        x2, y2 = divmod(goal.index(i), 3)
        distance += abs(x1 - x2) + abs(y1 - y2)
    return distance

# A* algorithm
def astar(start, goal):
    pq = []
    heapq.heappush(pq, (0 + manhattan_distance(start, goal), 0, [start]))
    visited = set()

    while pq:
        _, cost, path = heapq.heappop(pq)
        state = path[-1]

        if state == goal:
            return path

        visited.add(tuple(state))

        for move in get_moves(state):
            if tuple(move) not in visited:
                new_cost = cost + 1
                priority = new_cost + manhattan_distance(move, goal)
                heapq.heappush(pq, (priority, new_cost, path + [move]))

# ---- MAIN PROGRAM ----
start_state = [1, 2, 3,
               4, 0, 6,
               7, 5, 8]

goal_state = [1, 2, 3,
              4, 5, 6,
              7, 8, 0]

print("=== A* Algorithm (with Manhattan Distance) ===")
print("Start State:")
print_puzzle(start_state)
print("Goal State:")
print_puzzle(goal_state)

result = astar(start_state, goal_state)

print("Number of moves:", len(result)-1)
print("Sequence of moves:")
for step in result:
    print_puzzle(step)
```
